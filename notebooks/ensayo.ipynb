```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/anfagudelogo-tpt/datasets/refs/heads/main/car_price_dataset.csv')
```

```python
#librerias
import matplotlib.pyplot as plt 
import seaborn as sns
```

Procesamiento de datos

```python

df.info()
print(df.head())
```

#Calidad. Nulos y duplicados.

```python
#Duplicados
print(df.duplicated().sum())
#no hay duplicados

#Nulos 
print(df.isna().mean())
#no hay nulos 
```

Generación de gráficos

```python
#Precio por marca
#Aislar los datos
df.columns #'Brand', 'Model', 'Year', 'Engine_Size', 'Fuel_Type', 'Transmission', 'Mileage', 'Doors', 'Owner_Count', 'Price']
datos_precio_marca = df[["Price", "Brand"]]

#plot
plt.figure(figsize=(12, 9))
sns.boxplot(x = "Brand", y= "Price", data = datos_precio_marca )
plt.title("Distribución de precios por marca")
plt.xlabel("Marca")
plt.ylabel("Precio")
plt.xticks(rotation = 45)
plt.show()

#
```
